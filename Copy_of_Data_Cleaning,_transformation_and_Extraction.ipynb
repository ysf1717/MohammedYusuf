## Handling Missing Values


```python
# get to know % of missing values
def percent_missing(df):

  # Calculate the total number of cell in my dataframe
  totalCells = np.product(df.shape)

  # count the number of missing values per column
  missingCount = df.isnull().sum()

  # Calculate the total number of missing values
  totalMissing = missingCount.sum()

  # calculate the percemmtage of missing
  print("the Diabetes dataset contains", round(((totalMissing/totalCells) * 100), 2), "%", "missing Values.")

percent_missing(db)
```
